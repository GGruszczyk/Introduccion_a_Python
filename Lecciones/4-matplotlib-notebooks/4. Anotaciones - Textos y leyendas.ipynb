<h1>Anotaciones - Textos y leyendas</h1>

Como se ha podido ver, hasta el momento los gráficos realizados tienen algunas carencias claras:

<ul>
<li>No indican el significado de cada uno de los ejes.</li>
<li>No disponen de una leyenda que explique el significado de los diferentes colores/series incluidos.</li>
<li>En algunos casos, los ticks no representan los valores que se quieren representar.</li>
</ul>

En este notebook vamos a ver las alternativas que nos ofrece matplotlib para incluir en nuestros gráficos estos elementos "textuales" de forma que sean completamente autoexplicativos.

Para empezar, como siempre, importamos el módulo y cargamos los sets de datos.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mtcars = pd.read_table("data/mtcars.txt", sep = ",")
diamonds = pd.read_table("data/diamonds.txt", sep = ",")
```

<h2>Nombres en ejes y título del gráfico</h2>

Para incluir identificaciones en los ejes y en el propio gráfico, matplotlib pone a nuestra disposición tres funciones bien diferenciadas:

<ul>
<li><b>title:</b> Título del gráfico.</li>
<li><b>xlabel:</b> Etiqueta del eje X.</li>
<li><b>ylabel:</b> Etiqueta del eje Y.</li>
</ul>

```python
# Título del gráfico y Etiquetas de ejes
cuts = list(diamonds['cut'].unique())
x_values = [cuts.index(element) + np.random.uniform(0,0.6) 
            for element in diamonds['cut']]
y_values = diamonds['price']

plt.scatter(x_values, y_values, color='b', alpha=.1)
plt.xlim(-.5,5)
plt.ylim(-.5,20000)
plt.title('Price vs Cut', weight='bold', size=15, color='.3') 
plt.xlabel('Cut', weight='bold', size=12)
plt.ylabel('Price', weight='bold', size=12)

plt.show()
```

<h2>Ticks de los ejes</h2>

Aunque los gráficos de matplotlib necesitan valores numéricos en ambos ejes, en ocasiones dichos valores numéricos hacen referencia a valores categóricos (variables discretas) y es necesario modificar las etiquetas de los valores asignados en los ejes.

Para poder llevar a cabo esta modificación disponemos de las funciones:

<ul>
<li><b>xticks:</b> Para modificar los ticks del eje X</li>
<li><b>yticks:</b> Para modificar los ticks del eje Y</li>
</ul>

```python
# Etiquetas de ticks en eje X y Etiquetas de ticks en eje Y
plt.scatter(x_values, y_values, color='b', alpha=.1)
plt.xlim(-.5,5)
plt.ylim(-.5,20000)
plt.title('Price vs Cut', weight='bold', size=15, color='.3')
plt.xlabel('Cut', weight='bold', size=12)
plt.ylabel('Price', weight='bold', size=12)
plt.xticks(range(0, 5), cuts, rotation=45)
plt.yticks(range(0,20000,2500))
plt.show()
```

<h2>Leyendas</h2>

Al contrario que en ggplot2, en matplotlib es necesario incluir las leyendas de forma manual ya que los gráficos "por defecto" no las incluyen.

matplotlib pone a nuestra disposición la función <b>legend</b> para el renderizado de leyendas y la gestión de su posicionamiento, contenido, etc.

```python
# leyenda automática desde el contenido del gráfico
plt.scatter(mtcars[mtcars.am==0].mpg, mtcars[mtcars.am==0].hp, color=".5", 
            marker = (5, 1, 0), s=100, label='Automatic')
plt.scatter(mtcars[mtcars.am==1].mpg, mtcars[mtcars.am==1].hp, color="0",
            marker=(4, 1, 0), s=100, label='Manual')
plt.legend()
plt.show()
```

```python
# Establecimiento manual de contenido
serie_1 = plt.scatter(mtcars[mtcars.am == 0].mpg, mtcars[mtcars.am == 0].hp, 
                      color = "0.5", marker = (5, 1, 0), s = 100)
serie_2 = plt.scatter(mtcars[mtcars.am == 1].mpg, mtcars[mtcars.am == 1].hp, 
                      color = "0.0", marker = (4, 1, 0), s = 100.)
plt.legend((serie_1, serie_2), ('Automatic', 'Manual'))
plt.show()
```

```python
# Posicionamiento manual
serie_1 = plt.scatter(mtcars[mtcars.am == 0].mpg, mtcars[mtcars.am == 0].hp, 
                      color = "0.5", marker = (5, 1, 0), s = 100)
serie_2 = plt.scatter(mtcars[mtcars.am == 1].mpg, mtcars[mtcars.am == 1].hp, 
                      color = "0.0", marker = (4, 1, 0), s = 100.)
plt.legend((serie_1, serie_2), ('Automatic', 'Manual'), loc='upper center')
plt.show()
```

```python
# Disposición de elementos de la leyenda
serie_1 = plt.scatter(mtcars[mtcars.am == 0].mpg, mtcars[mtcars.am == 0].hp,
                      color = "0.5", marker = (5, 1, 0), s = 100.)
serie_2 = plt.scatter(mtcars[mtcars.am == 1].mpg, mtcars[mtcars.am == 1].hp,
                      color = "0.0", marker = (4, 1, 0), s = 100.)
plt.legend((serie_1, serie_2), ('Automatic', 'Manual'), loc = "lower left",
           ncol = 2)
plt.show()
```

```python
# Título para la leyenda
serie_1 = plt.scatter(mtcars[mtcars.am == 0].mpg, mtcars[mtcars.am == 0].hp,
                      color = "0.5", marker = (5, 1, 0), s = 100.)
serie_2 = plt.scatter(mtcars[mtcars.am == 1].mpg, mtcars[mtcars.am == 1].hp, 
                      color = "0.0", marker = (4, 1, 0), s = 100.)
plt.legend((serie_1, serie_2), ('Automatic', 'Manual'), title = "Gear type")
plt.show()
```
